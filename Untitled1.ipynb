```python
!pip install dateparser
import pandas as pd
import dateparser

# Creating a sample df_master for demonstration (add this if df_master is not defined elsewhere)
data = {
    'Faltu_ID': [1, 2, 3, 4, 5],
    'Full_Name': ['Alice Smith', 'Bob Johnson', 'Charlie Brown', 'Diana Prince', 'Eve Adams'],
    'Join_Date': ['2021-01-15', '2022-03-20', '2021-11-01', '2023-05-10', '2022-08-25'],
    'Salary': ['$50000', '$65000', '$72000', '$80000', '$55000']
}
df_master = pd.DataFrame(data)

# 1. Start with df_master and drop 'Faltu_ID' to create df_clean
df_clean = df_master.drop(columns=['Faltu_ID']).copy()

# 2. Date aur Salary ki cleaning
df_clean['Join_Date'] = pd.to_datetime(df_clean['Join_Date'].apply(lambda x: dateparser.parse(str(x)))).dt.date
df_clean['Salary'] = df_clean['Salary'].astype(str).str.replace('$', '', regex=False)
df_clean['Salary'] = pd.to_numeric(df_clean['Salary'], errors='coerce')

# 3. Missing data (NaN) wali row khatam karein
df_clean.dropna(subset=['Salary'], inplace=True)


# Result dekhein
df_clean
```

```python
# Sort the DataFrame by 'Join_Date'
df_sorted_by_date = df_clean.sort_values(by='Join_Date')

# Display the sorted DataFrame
display(df_sorted_by_date)
```

# Task
Generate a bar chart to visualize employee salaries using 'Full_Name' and 'Salary' from `df_clean`, a pie chart showing the distribution of salaries from `df_clean`, and a line graph to display the salary trend over 'Join_Date' using `df_sorted_by_date`. Finally, display all generated charts.

## Generate Bar Chart for Salaries

### Subtask:
Create a bar chart to visualize the salary of each employee from the 'df_clean' DataFrame, using 'Full_Name' for the x-axis and 'Salary' for the y-axis. Ensure proper labels and title are included.


**Reasoning**:
To visualize the salaries, I will create a bar chart using `matplotlib.pyplot`. This involves importing the library, plotting the data from `df_clean` with 'Full_Name' on the x-axis and 'Salary' on the y-axis, and then adding appropriate labels and a title for clarity.



```python
import matplotlib.pyplot as plt

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(df_clean['Full_Name'], df_clean['Salary'], color='skyblue')

# Label the x-axis
plt.xlabel('Employee Name')

# Label the y-axis
plt.ylabel('Salary')

# Add a title to the chart
plt.title('Employee Salaries')

# Rotate x-axis labels for better readability if names are long
plt.xticks(rotation=45, ha='right')

# Ensure layout is tight to prevent labels from overlapping
plt.tight_layout()

# Display the chart
plt.show()
```

## Generate Pie Chart for Salary Distribution

### Subtask:
Create a pie chart to show the distribution of salaries among employees using data from the 'df_clean' DataFrame. The chart should display percentages for each employee's salary contribution to the total.


**Reasoning**:
To create a pie chart showing salary distribution with percentages, I need to calculate the percentage of each employee's salary relative to the total salary and then plot it using matplotlib.pyplot.



```python
total_salary = df_clean['Salary'].sum()
df_clean['Salary_Percentage'] = (df_clean['Salary'] / total_salary) * 100

plt.figure(figsize=(10, 8))
plt.pie(
    df_clean['Salary_Percentage'],
    labels=df_clean['Full_Name'],
    autopct=lambda p: '{:.1f}%'.format(p) if p > 0 else '', # Display percentage only if greater than 0
    startangle=90,
    pctdistance=0.85
)
plt.title('Salary Distribution Among Employees')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

## Generate Line Graph for Salary Trend

### Subtask:
Create a line graph to visualize the trend of 'Salary' over 'Join_Date' using the 'df_sorted_by_date' DataFrame. This will show how salaries are distributed over time. Include appropriate labels, a title, and markers for clarity.


**Reasoning**:
To visualize the salary trend over time, I will create a line graph using `matplotlib.pyplot`. This involves plotting 'Join_Date' against 'Salary' from the `df_sorted_by_date` DataFrame, and then adding relevant labels, a title, and markers for better readability.



```python
import matplotlib.pyplot as plt

# Create a line graph
plt.figure(figsize=(12, 6))
plt.plot(df_sorted_by_date['Join_Date'], df_sorted_by_date['Salary'], marker='o', linestyle='-')

# Add a title to the chart
plt.title('Salary Trend Over Join Date')

# Label the x-axis
plt.xlabel('Join Date')

# Label the y-axis
plt.ylabel('Salary')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Ensure layout is tight to prevent labels from overlapping
plt.tight_layout()

# Display the chart
plt.show()
```
